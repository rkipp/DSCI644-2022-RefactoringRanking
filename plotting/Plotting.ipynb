```python
from scipy.io.arff import loadarff
import pandas as pd
import matplotlib.pyplot as plt
import io
from os import listdir
from os.path import isfile, join
%matplotlib inline
files = [f for f in listdir('./') if isfile(join('./', f))]

plt.style.use('seaborn-whitegrid')
```

The following code is used to plot the precision and recall for each of the response variables.

It does so using the .arff files, retrieved from the weka model.

The next two blocks compare the adangel and hibernate projects. The last two block compare the three models on the adangel project.

```python
clfs = ['J48', 'NB', 'RF']
cats = ['ExtremelyHigh', 'High', 'Low', 'Medium']
xcol = 'Recall'
ycol = 'Precision'
for cat in range(0,4):
    j48 = f'ROC_PC_Data_Hibernate_model_{cats[cat]}.arff'
    jdf = pd.DataFrame(loadarff(j48)[0])
    nb = f'DS3_RF_{cats[cat]}_ROC.arff'
    ndf = pd.DataFrame(loadarff(nb)[0])
    fig = plt.figure(figsize=(5,4))
    ax = fig.gca()
    ax.plot(jdf[xcol], jdf[ycol], color='orange', label='hibernate', linewidth=3)
    ax.plot(ndf[xcol], ndf[ycol], color='steelblue', label='adangel/pmd', linewidth=3)
    #plt.legend(loc= 'center left')
    #plt.title(cats[cat])
    plt.xlabel(xcol)
    plt.ylabel(ycol)
    plt.axhline(0, color='slategrey')
    plt.axvline(0, color='slategrey')
    plt.savefig('./Output/COMPARISON_ADANGEL_HIBERNATE'+cats[cat]+'_PR_Curve.png', bbox_inches='tight')
```

```python
clfs = ['J48', 'NB', 'RF']
cats = ['ExtremelyHigh', 'High', 'Low', 'Medium']
xcol = 'False Positive Rate'
ycol = 'True Positive Rate'
for cat in range(0,4):
    j48 = f'ROC_PC_Data_Hibernate_model_{cats[cat]}.arff'
    jdf = pd.DataFrame(loadarff(j48)[0])
    nb = f'DS3_RF_{cats[cat]}_ROC.arff'
    ndf = pd.DataFrame(loadarff(nb)[0])
    fig = plt.figure(figsize=(5,4))
    ax = fig.gca()
    ax.plot(jdf[xcol], jdf[ycol], color='orange', label='hibernate', linewidth=3)
    ax.plot(ndf[xcol], ndf[ycol], color='steelblue', label='adangel/pmd', linewidth=3)
    #plt.legend(loc= 'center left')
    #plt.title(cats[cat])
    plt.xlabel(xcol)
    plt.ylabel(ycol)
    plt.axhline(0, color='slategrey')
    plt.axvline(0, color='slategrey')
    plt.savefig('./Output/COMPARISON_ADANGEL_HIBERNATE'+cats[cat]+'_ROC_Curve.png', bbox_inches='tight')
```

```python
clfs = ['J48', 'NB', 'RF']
cats = ['ExtremelyHigh', 'High', 'Low', 'Medium']
xcol = 'False Positive Rate'
ycol = 'True Positive Rate'
for cat in range(0,4):
    j48 = [x for x in files if 'J48_'+cats[cat] in x]
    jdf = pd.DataFrame(loadarff(j48[0])[0])
    nb = [x for x in files if 'NB_'+cats[cat] in x]
    ndf = pd.DataFrame(loadarff(nb[0])[0])
    rf = [x for x in files if 'RF_'+cats[cat] in x]
    rdf = pd.DataFrame(loadarff(rf[0])[0])
    fig = plt.figure(figsize=(5,4))
    ax = fig.gca()
    ax.plot(jdf[xcol], jdf[ycol], color='purple', label='J48 Decision Tree', linewidth=3)
    ax.plot(ndf[xcol], ndf[ycol], color='seagreen', label='Naive Bayes', linewidth=3)
    ax.plot(rdf[xcol], rdf[ycol], color='orange', label='Random Forest', linewidth=3)
    #plt.title(cats[cat])
    plt.xlabel(xcol)
    plt.ylabel(ycol)
    plt.axhline(0, color='slategrey')
    plt.axvline(0, color='slategrey')
    plt.savefig('./Output/'+cats[cat]+'_ROC_Curve.png', bbox_inches='tight')
```
